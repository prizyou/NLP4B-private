# Language Modeling
Nowadays advances in the field of NLP use a couple of methodological approaches that allow to process large-scale datasets efficiently while losing as little information on textual data as possible. These methods include:

- tokenization
- embedding
- self-attention

Today, we will look at the first step, tokenization. Tokenization is the process of splitting a piece of text into smaller units, which are called tokens. Tokens can be words, numbers, or punctuation marks and are then usually represented by an integer. E.g., the word "house" then get's the integer 5837, "football" the integer 73. In general, there is no clear order, so integer 81 and 82 can represent completely different things from a semantical perspective.

Tokenization is fundamental as it is the first step in converting text into data that a machine can process. There are various ways in which text can be split into tokens without a clear winner. Let's start with the simplest.

## Word Level Tokenization
Word level tokenization, just as the term suggests, represents one word as one token (integer). Whitespace based splitting is the most trivial approach with many shortcomings. Nevertheless, it is the easiest, so let's have a look at it.

````python
# A simple example of tokenization
text = "Hello, world! Let's explore the concept of tokenization."

# Using Python's built-in method split()
tokens = text.split()

print(tokens)
````

In the above Python code, the split() method is used to tokenize a string into words. This is a very simple form of tokenization.

However, more sophisticated methods are available that can deal with the complexities of language. For instance, punctuation marks are still part of tokens, and words like "Let's" are not split into "Let" and "'s". To address this, NLP libraries such as NLTK or spaCy can be used.

````python
# Tokenization with NLTK
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Downloading the necessary datasets

text = "Hello, world! Let's explore the concept of tokenization."
tokens = word_tokenize(text)

print(tokens)
````

The word_tokenize method from NLTK provides a more refined tokenization that handles punctuation and splitting of contractions.

## Character Level Tokenization

Character-level tokenization involves breaking down text into individual characters. This method treats every character as a separate token and is useful when you want to train a model that learns the structure of words from the ground up, character by character.

This approach can be particularly beneficial when dealing with languages that have many compound words or when working with morphologically rich languages where the meaning of words can be altered with the addition of prefixes or suffixes. It's also useful in tasks like spelling correction where the model needs to understand text at the granularity of characters.

Here's a brief description followed by a simple Python code snippet that performs character-level tokenization:

Description:

Character-level tokenization simplifies the input data to the most granular level. Unlike word-level tokenization, which may need to deal with a large vocabulary, character tokenization works with a much smaller set of possible tokens. The trade-off, however, is that the sequence length increases, as each character becomes a token in itself.

````python
# Example of character level tokenization
text = "Hello, world!"

# Convert text into a list of characters
char_tokens = list(text)

print(char_tokens)
````

In the example above, the string "Hello, world!" is broken down into its constituent characters. Each character, including the comma and the space, is treated as an individual token.

This output of character-level tokenization would be:

````python
['H', 'e', 'l', 'l', 'o', ',', ' ', 'w', 'o', 'r', 'l', 'd', '!']
````

If you were to apply this to a dataset, you might store the list of unique characters as your vocabulary. For instance:

````python
# Define vocabulary for character-level tokenization
unique_chars = set(char_tokens)

print(unique_chars)
````

The unique_chars variable will contain all unique characters that the model needs to know.

Keep in mind that in real-world scenarios, the text should be preprocessed to include only the characters that are meaningful for the specific task, and the vocabulary should include all characters from the dataset, not just a single string.

##Advanced Tokenization Techniques

While basic tokenization can split text into words, it often fails to capture the semantic relationship between them. Advanced techniques such as subword tokenization and byte pair encoding (BPE) are used in modern NLP models.

Byte Pair Encoding (BPE):

BPE is a hybrid between word-level and character-level tokenization. It starts with a large corpus of text and gradually learns a fixed-size vocabulary of the most frequent and thus informative byte pairs, combining them into single tokens.
